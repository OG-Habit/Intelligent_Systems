```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Generating random data in a normal distribution

```python
np.random.seed(239)
x = np.linspace(0, 50, 100)
y = 2 * x + 1 + np.random.normal(0, 1, 100)
```

Creating a scatter-plot graph

```python
plt.figure(figsize=(8,6))
plt.scatter(x, y, label='Simulated Data', color='blue', marker='o')
plt.title('Simulated Data Scatter Plot')
plt.xlabel('Dependent Variable')
plt.ylabel('Independent Variable')
plt.legend()
plt.grid(True)
plt.show()
```
